## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
ofmicenmet = pd.merge(study_results, mouse_metadata, on='Mouse ID')

# Display the data table for preview
ofmicenmet
```

```python
# Checking the number of mice.
micenum= len(ofmicenmet["Mouse ID"].unique())
micenum
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
micedup= ofmicenmet[ofmicenmet.duplicated(subset= ["Mouse ID", "Timepoint"])]
micedup
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
ofmicenmet[ofmicenmet["Mouse ID"]== "g989"]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
miceclean= ofmicenmet[ofmicenmet["Mouse ID"]!= "g989"]

miceclean
```

```python
# Checking the number of mice in the clean DataFrame.
micecleannum= len(miceclean["Mouse ID"].unique())
micecleannum
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

micegroupy = miceclean.groupby(["Drug Regimen"])

micemean= micegroupy['Tumor Volume (mm3)'].mean()
micemedian= micegroupy['Tumor Volume (mm3)'].median()
micevar= micegroupy['Tumor Volume (mm3)'].var()
micestd= micegroupy['Tumor Volume (mm3)'].std()
micesem= micegroupy['Tumor Volume (mm3)'].sem()


# This method is the most straighforward, creating multiple series and putting them all together at the end.
mice_df= pd.DataFrame({"mean" : micemean,
                      "median": micemedian,
                      "variance": micevar,
                      "Standard Deviation": micestd,
                      "SEM": micesem
                      })
mice_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function
miceclean.groupby(["Drug Regimen"]).agg({'Tumor Volume (mm3)': ["mean", "median", "var", "std", "sem"]})
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

#make list of drugs to find from df
drugfours= ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
# Start by getting the last (greatest) timepoint for each mouse

micesort= miceclean.sort_values(by= ["Timepoint"])
timetime= micesort.groupby(["Mouse ID", "Drug Regimen"]).last()
timetime.reset_index(inplace= True)
drugtimes = timetime.loc[timetime["Drug Regimen"].isin(drugfours)]
drugtimes

```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugfours= ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
finaldrugs= []

#for loop to go through each drug
for drugs in drugfours:
    finaltreatment= drugtimes.loc[drugtimes["Drug Regimen"]== drugs, "Tumor Volume (mm3)"]
    finaldrugs.append(finaltreatment)

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
#calculate other stuff for quartiling
    quartiles = finaltreatment.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
#print stufff 
    print(drugs)
    print(finaltreatment[(finaltreatment<lower_bound) | (finaltreatment>upper_bound)])
    print(f"lower quartile of tumor volume: {lowerq}")
    print(f"upper quartile of tumor volume : {upperq}")
    print(f"interquartile range of tumor volume : {iqr}")
    print(f"Values below {lower_bound} might be outliers.")
    print(f"Values above {upper_bound} might be outliers.")

    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

fig1, ax1 = plt.subplots()
ax1.set_title("Mice Tumor Levels")
ax1.set_ylabel("tumor volumes")

#use 2 lists from previous cell on boxplot!
ax1.boxplot(finaldrugs, labels= drugfours)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
#use .loc to locate that mice

#specify a mouse treated with capomolin
capomolinmice = miceclean.loc[miceclean["Mouse ID"]=="y793"]

#plot
#dataframe no need for line, it's basic
capomolinmicedata = capomolinmice.plot(x="Timepoint",
              y= "Tumor Volume (mm3)",
              title= "Mice y793 on Capomolin Treatment"
             )
capomolinmicedata.set_xlabel("Timepoint(Months)")
capomolinmicedata.set_ylabel("Tumor Volume")
     
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
micencapomolin = miceclean.loc[miceclean["Drug Regimen"] =="Capomulin"]

#group mouse id
miceid= micencapomolin.groupby("Mouse ID")

#grab weight and tumor from the grouped df .mean
miceweight = miceid["Weight (g)"].mean()
micetumor = miceid["Tumor Volume (mm3)"].mean()

#make dataframe
miceplot = pd.DataFrame({
    "Tumor Volume (mm3)": micetumor,
    "Weight (g)": miceweight
})

#plot
miceplot.plot(kind = "scatter", x = "Weight (g)", y = "Tumor Volume (mm3)", title = "Mice Weight and Tumor Volume for Capomulin")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
#correlation
micecorrelate = st.pearsonr(miceplot["Tumor Volume (mm3)"],miceplot["Weight (g)"])
micecorrelate
print(f"The correlation coefficient between mice weight and tumor volume is is {round(micecorrelate[0],3)}")

# mouse weight and average tumor volume for the Capomulin regimen
#same 2nd line scatter
```

```python
#regression
from scipy.stats import linregress

x_values = miceplot['Weight (g)']
y_values = miceplot['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
